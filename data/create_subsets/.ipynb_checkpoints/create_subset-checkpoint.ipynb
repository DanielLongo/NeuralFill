```python
def get_full_filenames(filename, source):
    if source == "LPCH_NK":
        source_dir = "lpch/"
    elif source == "STANFORD_NK":
        source_dir = "stanford/"
    else:
        print("invalid source")
        return None
    
    filename = filename.split(".")[0] + ".eeghdf"
    return DATASET_PATH + source_dir + filename
```
